```python
from jamo import h2j, j2hcj


def get_jongsung_TF(sample_text):
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1]
    last_word_jamo_list = list(j2hcj(h2j(last_word)))
    last_jamo = last_word_jamo_list[-1]

    jongsung_TF = "T"

    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F"

    return jongsung_TF
```

```python
def make_user_dic_csv(morpheme_type, word_list, user_dic_file_name):
  file_data = []

  for word, score in word_list:
    jongsung_TF = get_jongsung_TF(word)

    line = f"{word},,,{score},{morpheme_type},*,{jongsung_TF},{word},*,*,*,*,*\n"

    file_data.append(line)

  with open("./user-dic/user-nnp.csv", 'w', encoding='utf-8') as f:
    for line in file_data:
      f.write(line)
```

```python
word_list = [('전자정부', 0), ('공공 마이데이터', 0)]

make_user_dic_csv(morpheme_type="NNP", word_list=word_list, user_dic_file_name='user-nnp.csv')
```

```python
import pandas as pd
import re
from collections import Counter
```

```python
df = pd.read_excel("/content/drive/MyDrive/godo/data.xls")
```

```python
# 제목 컬럼에서 특수문자 제거
df['제목'] = df['제목'].apply(lambda x: re.sub(r'[\(\)]', '', x))
df['제목'] = df['제목'].apply(lambda x: re.sub(r'\[[^\]]+\]', '', x))


# 제목 컬럼만 추출
titles = df['제목'].tolist()
```

```python

from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# CountVectorizer를 사용하여 2-gram부터 3-gram까지 추출
vectorizer = CountVectorizer(ngram_range=(2, 3), tokenizer=mecab.morphs)
X = vectorizer.fit_transform(titles)
ngram_array = X.toarray()
ngram_names = vectorizer.get_feature_names_out()

# 빈도수 계산
ngram_sums = np.sum(ngram_array, axis=0)
sorted_indices = np.argsort(ngram_sums)[::-1]

# 상위 n-gram과 빈도를 DataFrame으로 저장
ngram_df = pd.DataFrame({
    'ngram': np.array(ngram_names)[sorted_indices],
    'count': ngram_sums[sorted_indices]
})
```

```python
plt.rc('font', family='NanumBarunGothic')
```

```python
# 상위 10개 n-gram 빈도 막대그래프
top_n = 10
top_ngrams = ngram_df.head(top_n)
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='ngram', data=top_ngrams, orient='h')
plt.title('Top {} Frequent N-grams (2-gram to 3-gram)'.format(top_n))
plt.show()
```

```python
# '협약'이 포함된 제목만 필터링
hyeop_yak_titles = [title for title in titles if '협약' in title]

# 해당 제목에서 형태소 추출
hyeop_yak_words = []
for title in hyeop_yak_titles:
    hyeop_yak_words.extend(mecab.morphs(title))

# 형태소의 빈도수 계산
hyeop_yak_word_count = Counter(hyeop_yak_words)

# 상위 10개 형태소의 빈도
top_n = 10
top_hyeop_yak_words = hyeop_yak_word_count.most_common(top_n)
top_hyeop_yak_df = pd.DataFrame(top_hyeop_yak_words, columns=['morpheme', 'count'])

# 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='morpheme', data=top_hyeop_yak_df, orient='h')
plt.title('Top 10 Morphemes in Titles Containing "협약"')
plt.xlabel('Frequency')
plt.ylabel('Morpheme')
plt.show()
```

```python
from collections import defaultdict

def co_occurrence(sentences, window_size=2):
    d = defaultdict(int)
    vocab = set()
    for text in sentences:
        # 텍스트를 형태소로 분리
        text = mecab.morphs(text)
        vocab.update(text)
        for i, word in enumerate(text):
            for j in range(i-window_size, i+window_size+1):
                if 0 <= j < len(text) and j != i:
                    key = tuple(sorted([word, text[j]]))
                    d[key] += 1
    return d, vocab

sentences = df['제목'].tolist()
co_dict, _ = co_occurrence(sentences)
```

```python
target_word = '협약'  # 예시로 '협약'을 사용
co_words = defaultdict(int)
for key, value in co_dict.items():
    if target_word in key:
        if key[0] == target_word:
            co_words[key[1]] += value
        else:
            co_words[key[0]] += value

# 상위 10개 형태소를 DataFrame으로 변환
top_co_words = dict(sorted(co_words.items(), key=lambda item: item[1], reverse=True)[:10])
co_df = pd.DataFrame(list(top_co_words.items()), columns=['morpheme', 'co_count'])
```

```python
# 데이터 설정
labels = co_df['morpheme'].tolist()
sizes = co_df['co_count'].tolist()
colors = sns.color_palette('pastel', len(labels)).as_hex()
explode = [0.1] * len(labels)  # 각 조각을 약간 떼어내기 위한 설정

# 도넛 차트 그리기
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90, pctdistance=0.85,
        wedgeprops=dict(width=0.3))

# 중앙에 원형 흰색 원 그리기 (도넛 모양 만들기)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  # 원형을 유지하기 위한 설정
plt.title(f'Morphemes Co-occurring with "{target_word}"')
plt.show()
```
